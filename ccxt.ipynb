## login

```python
import ccxt

api_key = ''
api_secret = ''
exchange = ccxt.binance({
    'apiKey': api_key ,
    'secret': api_secret,
    'options': {
        'adjustForTimeDifference': True,
        'defaultType': 'future',
    },
})
```

## kbars

```python
import pandas as pd
symbol ='BTC/USDT'
ohlcv = exchange.fetch_ohlcv(symbol,'5m')
klines =pd.DataFrame(ohlcv,columns = ['Date', 'Open','High','Low','Close','Volume'])
klines['time'] = pd.to_datetime(klines['Date'],unit='ms')
klines.set_index('time', inplace=True)
klines
```

## fetch_Balance

```python
balance = float(exchange.fetch_balance()['info']['availableBalance'])
balance 
```
